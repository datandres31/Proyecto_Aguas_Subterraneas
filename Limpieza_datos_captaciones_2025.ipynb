# GEOREFERENCIACIÓN DE LOS PUNTOS DE CAPTACIÓN.

```python
def dms_to_dd(grados, minutos, segundos):
    # Ojo: aquí se conserva el signo de 'grados' (ej. -74 ya trae negativo)
    dd = abs(grados) + (minutos / 60) + (segundos / 3600)
    return -dd if grados < 0 else dd
```

```python
puntos_captacion_sabana.loc[:, "LAT_DD_calc"] = puntos_captacion_sabana.apply(lambda row: dms_to_dd(row["GRADOS_LAT"], row["MINUTOS_LAT"], row["SEGUNDOS_LAT"]), axis=1)
puntos_captacion_sabana.loc[:, "LONG_DD_calc"] = puntos_captacion_sabana.apply(lambda row: dms_to_dd(row["GRADOS_LONG"], row["MINUTOS_LONG"], row["SEGUNDOS_LONG"]), axis=1)
```

# Fundamentación de la variable ratio caudal asignado/oferta disponible

Relación oferta–demanda en hidrogeología

El caudal asignado refleja la demanda reconocida y autorizada por la autoridad ambiental.

La oferta disponible corresponde a la recarga neta utilizable o el rendimiento sostenible del acuífero (según balance hídrico).

La relación entre ambos permite aproximarse al concepto de índice de explotación (también llamado índice de presión sobre el recurso en la literatura internacional).

Interpretación del índice

 menor a 1 → La demanda asignada es menor que la oferta → explotación potencialmente sostenible.

 igual a 1 → La demanda está al límite de la oferta → riesgo de sobreexplotación en escenarios de incertidumbre o variabilidad climática.

 menor a 1 → Se asignó más caudal del que la oferta permite → situación de sobreconcesión, lo cual con el tiempo puede conducir a abatimiento piezométrico, reducción de caudales base en ríos y deterioro de humedales.

Soporte conceptual

En estudios de gestión de acuíferos se usan métricas como el Groundwater Exploitation Index (GEI) o el índice de presión hídrica (Falkenmark, UNESCO, FAO).

Estos indicadores siempre relacionan extracción/autorización con oferta renovable, exactamente lo que esta variable plantea

# Función para definición de los ratio 

```python
import numpy as np
import pandas as pd

def calcular_ratios(df, col_concesionado, col_asignado, col_oferta_total, col_oferta_disp):
    """
    Calcula cuatro ratios clave en gestión hídrica subterránea:
    1. Asignado / Oferta disponible
    2. Concesionado / Oferta total
    3. Concesionado / Oferta disponible
    4. Asignado / Concesionado

    Manejo de ceros:
    - Si denominador = 0 y numerador > 0 → np.inf (sobreconcesión absoluta)
    - Si denominador = 0 y numerador = 0 → np.nan (no aplica)
    - Si denominador > 0 → ratio normal
    """

    def safe_div(num, den):
        if den == 0 and num > 0:
            return np.inf
        elif den == 0 and num == 0:
            return np.nan
        else:
            return num / den

    df = df.copy()

    df["RATIO_ASIGNADO_DISPONIBLE"] = df.apply(
        lambda row: safe_div(row[col_asignado], row[col_oferta_disp]), axis=1
    )

    df["RATIO_CONCESIONADO_TOTAL"] = df.apply(
        lambda row: safe_div(row[col_concesionado], row[col_oferta_total]), axis=1
    )

    df["RATIO_CONCESIONADO_DISPONIBLE"] = df.apply(
        lambda row: safe_div(row[col_concesionado], row[col_oferta_disp]), axis=1
    )

    df["RATIO_ASIGNADO_CONCESIONADO"] = df.apply(
        lambda row: safe_div(row[col_asignado], row[col_concesionado]), axis=1
    )

    # Reemplazar infinitos por NaN si quieres evitar que dañen estadísticas
    df.replace([np.inf, -np.inf], np.nan, inplace=True)

    return df
```

# INTERPRETACIÓN HIDROGEOLOGICA DE LOS RATIOS.


1. RATIO_ASIGNADO_DISPONIBLE

Máximo: 829,000 → valores extremos indican casos donde prácticamente se asignó agua en sitios sin oferta.

Interpretación hidrogeológica:

Muestra una sobreexplotación sistémica del acuífero, donde los caudales distribuidos superan con creces la capacidad de recarga utilizable.

Esto implica riesgos de descensos piezométricos, pérdida de caudal base en ríos y desecación de humedales, problemas ya documentados en la Sabana.


2. RATIO_ASIGNADO_CONCESIONADO

Mediana: 0.5 → en la mayoría de los casos los usuarios ejercen solo la mitad de lo que se les ha concesionado.

Moda: 1.0 → hay un número importante de usuarios que usan todo lo que les fue otorgado.

Interpretación hidrogeológica:

Hay un margen de subutilización en varias concesiones.

Esto refleja una ineficiencia administrativa: se han otorgado concesiones sobredimensionadas respecto al uso real, lo que aumenta artificialmente la presión legal sobre el recurso.


3. RATIO_CONCESIONADO_DISPONIBLE

Mediana: ~242 → la presión legal sobre la oferta disponible es todavía más crítica que la ejercida efectivamente.

Máximo: 864,000 → concesiones gigantes sobre zonas con oferta prácticamente nula.

Interpretación hidrogeológica:

El marco de concesiones está totalmente desalineado con la capacidad natural del acuífero.

Esto es una señal clara de sobreconcesión estructural, que genera un pasivo hídrico a futuro, incluso si hoy no todo el volumen se usa.

4. RATIO_CONCESIONADO_TOTAL

Mediana: ~180 y Media >3,000 → nuevamente, la presión sobre la oferta bruta es desproporcionada.

Interpretación hidrogeológica:

Si incluso frente a la oferta hídrica total (sin descontar reservas ambientales) los caudales concesionados la superan en cientos de veces, significa que el sistema de otorgamiento de derechos ha sido muy permisivo.

Esto refuerza que el acuífero está sobrecomprometido legalmente, aunque no siempre se use en la práctica.

5. Conclusiones hidrogeológicas

Sobreconcesión y sobreexplotación:

Lo concesionado y asignado superan la oferta disponible en órdenes de magnitud.

Esto confirma que el acuífero de la Sabana de Bogotá está bajo un riesgo alto de insostenibilidad.

Ineficiencia administrativa:

El hecho de que el ratio asignado/concesionado sea <1 muestra concesiones sobredimensionadas.

Desde la gestión, hay margen para ajustar concesiones a la demanda real.

Riesgo ambiental:

Los humedales, ríos base y ecosistemas dependientes del agua subterránea están en situación crítica si las extracciones continúan en este nivel.

Recomendación de gestión:

Priorizar zonas con ratios extremos (>100) para revisión inmediata.

Clasificar los puntos en categorías de riesgo (bajo, medio, alto, crítico).

Usar estos índices como herramientas de diagnóstico en planificación de concesiones y actualización del Plan de Manejo del Acuífero.

# creación de la variable categorica de riesgo con base en el caudal asignado- oferta disponible.


```python
def clasificar_riesgo(ratio):
    if pd.isna(ratio):
        return "No aplica"
    elif ratio < 1:
        return "Sostenible"
    elif ratio < 10:
        return "Presión alta"
    elif ratio < 100:
        return "Crítico"
    else:
        return "Colapso potencial"
```

```python
puntos_captacion_sabana["RIESGO_ASIG_DISP"] = puntos_captacion_sabana["RATIO_ASIGNADO_DISPONIBLE"].apply(clasificar_riesgo)
```

```python
puntos_captacion_sabana.RIESGO_ASIG_DISP.value_counts()
```

```python
municipios_riesgo = pd.crosstab(puntos_captacion_sabana["MUNICIPIO"], puntos_captacion_sabana["RIESGO_ASIG_DISP"])

# Ordenar por el número de críticos
municipios_riesgo = municipios_riesgo.sort_values("Colapso potencial", ascending=False)

municipios_riesgo["TOTAL"] = municipios_riesgo.sum(axis=1)

municipios_riesgo.loc["TOTAL_GENERAL"] = municipios_riesgo.sum()

municipios_riesgo
```

# Resultados de unicidad

IDCAPTACION (1132 únicos)
→ Son los puntos físicos de extracción (pozos, aljibes).
✅ Este es el universo real de captaciones en tu base.

IDCONCESION (1100 únicos)
→ Actos administrativos que autorizan la explotación.
🔎 Muy cercano al número de captaciones → en la práctica hay casi un 1 a 1, pero algunas captaciones tienen más de una concesión (modificaciones, renovaciones, prórrogas).

IDPREDIO (1091 únicos)
→ Predios asociados a los puntos.
📌 Casi igual al número de concesiones/captaciones → indica que cada predio suele tener una captación, aunque existen casos de:

Varios predios con la misma captación (uso compartido).

Un predio con varias captaciones.

IDUSO (1916 únicos)
→ Usos declarados: agrícola, pecuario, doméstico, industrial, etc.
🚨 Este número es mayor que captaciones/concesiones → significa que un mismo pozo o concesión puede tener múltiples usos asociados. Ejemplo: un pozo agrícola que también abastece consumo doméstico.

IDUSUARIO (834 únicos)
→ Personas naturales o jurídicas titulares de concesiones.
👀 Menor que captaciones/concesiones → significa que un mismo usuario concentra varias captaciones y concesiones. Esto es típico en:

Empresas con varios predios.

Distritos de riego o municipios.

Grandes usuarios (ej. industria, floricultores).

🧭 Interpretación estratégica

Concentración de uso

834 usuarios concentran 1132 captaciones → algunos usuarios manejan varios pozos.

Esto permite priorizar fiscalización: con pocos usuarios puedes cubrir gran parte de la presión sobre el acuífero.

Fragmentación de usos

El salto de 1132 captaciones a 1916 usos indica que la presión no es solo por número de pozos, sino por la diversidad de demandas sobre cada pozo.

Esto complica la planificación, porque un mismo punto abastece varios sectores.

Gestión por predios vs. concesiones

Dado que predios (1091) y concesiones (1100) están casi balanceados, puedes trabajar el análisis tanto desde la perspectiva predial como desde la jurídico-administrativa sin perder consistencia.

Implicación para georreferenciación

Tu nivel más estable es IDCAPTACION (1132 únicos).

Si georreferencias a nivel de concesión o uso, vas a duplicar puntos en el mapa.

Lo más sólido es georreferenciar en torno a la captación física y luego relacionar las tablas de concesiones, predios, usos y usuarios.
